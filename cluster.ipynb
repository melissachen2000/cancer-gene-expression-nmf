```python
#these values should be set based on the specifics of the dataset

clusters = 5  # number of clusters
radius = 2  # radius of the large circle that contains all 5 circles

#the length of this dictionary should be changed so that len(colors) = clusters
colors = {0:"yellow", 1:"green", 2:"blue", 3:"purple", 4:"red"}  # colors assigned to each point based on ground truth
```

```python
#set the values for predicted_labels and ground_truth
predicted_labels = [i % clusters for i in range(1000)]
ground_truth = [i % clusters for i in range(1000)]
```

```python
angle = 2 * math.pi / clusters  # angle between the center of any two clusters
```

```python
central_point = {}  # central point of the distinct circles, each cluster will have radius 1

for i in range(clusters):
    central_point[i] = (radius * math.cos(angle * i), radius * math.sin(angle * i))
```

```python
#assigning point coordinates and color
coords = {}
for i in range(len(predicted_labels)):
    a = random.random() * 2 * math.pi
    r = math.sqrt(random.random())
    
    (x,y) = central_point[predicted_labels[i]]
    
    coords[r * math.cos(a) + x, r * math.sin(a) + y] = colors[ground_truth[i]]
```

```python
keys = tuple(coords.keys())

for index in range(len(keys)):
    plt.scatter(keys[index][0], keys[index][1], color=coords[keys[index]], s=15)
```
