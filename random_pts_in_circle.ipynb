```python
import math
import random
import matplotlib.pyplot as plt
```

```python
radius = .5
x = []
y = []
```

```python
for i in range(1000):
    a = random.random() * 2 * math.pi
    r = radius * math.sqrt(random.random())

    for i in range(0,5):
        x.append(r * math.cos(a) + i)
        y.append(r * math.sin(a) + i)
```

```python
for i in range(len(x)):
    plt.scatter(x[i],y[i])
```
